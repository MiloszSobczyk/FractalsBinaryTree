
## Fraktal Drzewa Binarnego

### Opis teoretyczny:
Drzewo binarne jest fraktalem w postaci drzewa, w którym każdy węzeł ma co najwyżej dwa potomków, często określane jako lewe i prawe poddrzewo.

Jest to jeden z prostszych fraktali, który znajduje zastosowanie w matematyce oraz w wizualizacji przyrodniczych obiektów i zjawisk. Są to np. struktura gałęzi drzew, układ oddechoowy człowieka, naczynia krwionośne czy wyładowania elektryczne(tzw. Figury Lichtenberga).

Drzewo binarne w matematyce jest konstrukcją rekurencyjną, w której każda gałąź dzieli się na dwie mniejsze, z określoną zmianą kąta i długości. W wyniku tego powstaje figura o strukturze samopodobnej, co oznacza, że mniejsze części drzewa przypominają całą jego strukturę.

**Historia:**

Fraktal drzewa binarnego jako matematyczny koncept nie ma jednego konkretnego odkrywcy ani daty odkrycia. Jest on wynikiem badań nad strukturami rekurencyjnymi i samopodobnymi, które były badane przez wielu matematyków. Niemniej jednak, rozwój teorii fraktali, które obejmują drzewa binarne, można przypisać ogólnym badaniom nad fraktalami i geometrią samopodobną, które zostały spopularyzowane przez Benoîta Mandelbrota w latach 70. XX wieku.

**Znaczenie matematyczne:**

Podstawowe założenia fraktala drzewa binarnego były stosowane w matematyce, szczególnie teorii grafów oraz w algorytmach i wykorzystywanych w nich strukturach danych.

### Algorytm Generowania

1. **Inicjalizacja**: Zaczynamy od narysowania początkowej linii (korzenia drzewa) o określonej długości, wychodzącej z punktu startowego, pod ustalonym kątem względem poziomu.

2. **Rekurencja**: Dla każdej narysowanej linii:
   - Rysujemy dwie nowe linie będące "potomkami" tej linii.
   - Każda nowa linia zaczyna się w punkcie końcowym poprzedniej.
   - Nowe linie są rysowane pod określonymi kątami względem poprzedniej linii (np. 45 stopni w lewo i w prawo).
   - Długość nowych linii stanowi ułamek długości linii macierzystej (np. 0.6).
   - Powtarzamy ten krok dla każdej nowej linii, zmniejszając głębokość rekurencji o 1, aż osiągniemy głębokość 0.

### Równania

Nowe współrzędne punktu końcowego $(x_{\text{new}}, y_{\text{new}})$ linii można obliczyć na podstawie początkowych współrzędnych $(x, y)$, długości linii `length` i kąta `angle`, korzystając z następujących równań trygonometrycznych:

$$
x_{\text{new}} = x + (\text{length} \times \cos(\text{angle}))
$$

$$
y_{\text{new}} = y + (\text{length} \times \sin(\text{angle}))
$$

Gdzie:
- $(x, y)$: współrzędne początkowe linii.
- `length`: długość linii.
- `angle`: kąt, pod którym linia jest rysowana.


### Modyfikacje Algorytmu

- **`starting_angle`**:
   - Ten parametr określa początkowy kąt, pod którym jest rysowana pierwsza, główna gałąź (korzeń drzewa). Zwykle jest ustawiony na 90 stopni (lub \( \pi/2 \) radianów), co oznacza, że korzeń drzewa jest rysowany pionowo w górę. Zmiana tego kąta może wpłynąć na orientację całego drzewa, powodując jego nachylenie w lewo lub w prawo.

- **`angle_change`**:
   - Parametr ten kontroluje, o ile stopni zmienia się kąt przy rysowaniu każdej kolejnej gałęzi. Na przykład, jeśli wartość wynosi 30 stopni, każda gałąź będzie się rozgałęziać pod kątem 30 stopni w lewo i 30 stopni w prawo w stosunku do poprzedniej. Zwiększenie wartości `angle_change` powoduje, że drzewo staje się bardziej rozłożyste, a zmniejszenie - że gałęzie są bardziej zwarte.

- **`length_modifier`**:
   - Jest to współczynnik, który określa, o ile każda nowa gałąź ma być krótsza od swojej gałęzi macierzystej. Na przykład, wartość `length_modifier` równa 0.7 oznacza, że każda nowa gałąź będzie miała 70% długości poprzedniej. Zmniejszenie tego współczynnika powoduje, że gałęzie szybko stają się krótsze, co prowadzi do bardziej kompaktowego drzewa. Wyższe wartości sprawiają, że długości gałęzi zmniejszają się wolniej, co daje bardziej rozłożystą strukturę.

- **`starting_length`**:
   - Ten parametr ustala początkową długość korzenia drzewa. Od niego zależy wielkość całego drzewa. Zwiększenie `starting_length` powoduje, że całe drzewo jest większe, natomiast mniejsze wartości skutkują bardziej zwarto wyglądającym drzewem.

### Wpływ Modyfikacji

Zmiany w tych parametrach pozwalają generować różne wersje drzewa binarnego. Zwiększenie kąta i zmniejszenie długości gałęzi tworzy szerokie, rozłożyste drzewa, podczas gdy mniejsze kąty i długości gałęzi prowadzą do smuklejszych, bardziej pionowych struktur. Kombinowanie tych parametrów umożliwia eksperymentowanie z różnymi kształtami fraktali, co zwiększa różnorodność wizualną.
Nie wpływa to jednak na samopodobieństwo i rekurencyjność fraktala, co jest jego istotną cechą.

### Analiza i Wnioski

Na zakończenie projektu zespół przeprowadził analizę porównawczą klasycznego fraktala drzewa binarnego z jego zmodyfikowaną wersją.

- **Zachowane cechy fraktalne**:
   - **Samopodobieństwo**: Zarówno klasyczna, jak i zmodyfikowana wersja drzewa zachowują kluczową cechę fraktali, jaką jest samopodobieństwo. Każda gałąź drzewa rozdziela się na mniejsze gałęzie, które mają ten sam ogólny wzór co całość, co jest charakterystyczną właściwością fraktali.
   - **Rekurencyjność**: W obu wersjach algorytm bazuje na rekurencyjnym procesie podziału linii, co tworzy wielopoziomową strukturę, gdzie każda kolejna gałąź staje się punktem wyjścia dla następnych.

- **Najbardziej widoczne różnice**:
   - **Rozpiętość drzewa**: W zmodyfikowanej wersji drzewa, zmiana parametrów, takich jak `angle_change`, prowadzi do znacznie większej rozpiętości gałęzi, co sprawia, że drzewo wygląda na bardziej rozłożyste w porównaniu do klasycznej, bardziej zwartej wersji.
   - **Zmniejszająca się długość gałęzi**: Parametr `length_modifier` w zmodyfikowanej wersji powoduje szybsze lub wolniejsze kurczenie się długości gałęzi w porównaniu do oryginału. To wpływa na ogólny wygląd drzewa – zmodyfikowane drzewo może być bardziej kompaktowe lub wręcz przeciwnie, bardziej rozwinięte.
   - **Wielkość drzewa**: Wprowadzenie zmian do `starting_length` wpływa na całkowitą wysokość i rozmiar drzewa, co może skutkować zupełnie inną estetyką końcową w zmodyfikowanej wersji.

- **Wnioski z procesu implementacji i modyfikacji**:
   - **Wyzwania**: Głównym wyzwaniem było dobranie odpowiednich wartości dla parametrów modyfikacji, takich jak kąt rozgałęzień czy zmiennik długości. Zbyt duże lub zbyt małe wartości prowadziły do nieintuicyjnych wyników, co wymagało wielu prób, by uzyskać estetycznie wyglądające drzewo.
   - **Osiągnięcia**: Dzięki elastyczności algorytmu udało się uzyskać różnorodne warianty drzewa binarnego. Modyfikacje takie jak zmiana kąta rozgałęzień czy długości gałęzi pozwoliły na eksperymentowanie z różnymi formami, zachowując jednocześnie kluczowe właściwości fraktalne.

Proces modyfikacji algorytmu poszerzył nasze zrozumienie fraktali, a także pokazał, jak niewielkie zmiany parametrów mogą znacząco wpłynąć na końcową strukturę i wygląd generowanego obiektu.


```python

import matplotlib.pyplot as plt
import numpy as np
import imageio.v3
import os

def draw_branch(ax, x, y, depth, starting_angle, starting_length, angle_change, length_modifier):
    if depth == 0:
        return
    
    # Calculate branch endpoints
    x_end = x + starting_length * np.cos(starting_angle)
    y_end = y + starting_length * np.sin(starting_angle)
    
    # Draw branch
    ax.plot([x, x_end], [y, y_end], 'brown', lw=2)

    # Calculate new length and depth    
    new_length = starting_length * length_modifier
    new_depth = depth - 1

    # Calculate left and right angles    
    angle_left = starting_angle + angle_change
    angle_right = starting_angle - angle_change
    
    # Draw branches recursively
    draw_branch(ax, x_end, y_end, new_depth, angle_left, new_length, angle_change, length_modifier)
    draw_branch(ax, x_end, y_end, new_depth, angle_right, new_length, angle_change, length_modifier)

def draw_tree(depth, angle, length, angle_change, length_modifier):
    fig, ax = plt.subplots(figsize = (8, 8))
    ax.set_aspect('equal')
    ax.axis('off')
    
    # Starting point
    x, y = 0, 0
    
    # Draw tree with n levels
    draw_branch(ax, x, y, depth, angle, length, angle_change, length_modifier)
    
    return fig

# Parameters:
# path              - path for folder where to save files,
# gif_name          - generated gif name,
# delete_files      - specifies whether to delete intermediate files used for generating,
# iterations        - number of iterations used to generate fractal,
# starting_angle    - starting angle of a binary tree,
# starting_length   - starting length of a binary tree,
# angle_change      - change in branch angle with each iteration,
# length_modifier   - modifier in branch length in each iteration,
def create_tree_gif(path = './', gif_name = 'tree_animation.gif', delete_files = True, iterations = 6, starting_angle = np.pi / 2, starting_length = 1, angle_change = np.pi / 6, length_modifier = 0.7):
    filename = f'{path}{gif_name}';
    filenames = []
    for i in range(iterations + 1):
        fig = draw_tree(i, starting_angle, starting_length, angle_change, length_modifier)
        img_filename = f"{path}tree_{i}.png"
        fig.savefig(img_filename)
        plt.close(fig)
        filenames.append(img_filename)
    
    with imageio.get_writer(filename, mode = 'I') as writer:
        for img_filename in filenames:
            image = imageio.imread(img_filename)
            writer.append_data(image)
            if delete_files:
                os.remove(img_filename)

create_tree_gif('./gifs/')

```
